```python
import pandas as pd
from tensorflow.keras.models import load_model
import numpy as np
```

```python
model = load_model('/home/tomato/Desktop/UIT/DS105/model/model_temperature_TPHCM.h5')
df = pd.read_csv('/home/tomato/Desktop/UIT/DS105/Total_weather_dataset_in_HCM.csv')

def convertToMatrix(data, step):
    X, Y =[], []
    for i in range(len(data)-step):
        d=i+step  
        X.append(data[i:d,])
        Y.append(data[d,])
    return np.array(X), np.array(Y)
```
