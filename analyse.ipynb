```python
import pandas as pd
import os
```

```python
filelist = os.listdir('dataset')
# filelist[:] = 'dataset/' +filelist[:]
c = 0
for file in filelist:
    filelist[c] = 'dataset/' + file
    c+=1
print(filelist)
```

```python
filelist = sorted(filelist, key=os.path.getctime)
print(filelist)
```

```python
path = 'dataset/Weather_data_of_HCM_tan-binh_at_date_2022-11-03.csv'
df = pd.read_csv(path)
df2 = df.copy()
df2.columns = ['Time', 'Temperature', 'Dew Point', 'Humidity', 'Wind',
       'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.', 'Condition',' ']
print(df2.columns)
if 'Condition' in df2.columns:
    del df2['Condition']
if ' ' in df2.columns:
    del df2[' ']
df2['Date']=path[-14:-4]
df2.to_csv('temp.csv')
df2
```
